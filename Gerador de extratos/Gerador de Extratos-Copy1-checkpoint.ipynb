```python
import csv
import random
from datetime import datetime, timedelta

def random_date(year, month, start_day=1):
    start_date = datetime(year, month, start_day)
    end_date = datetime(year, month + 1, 1) if month < 12 else datetime(year + 1, 1, 1)
    return start_date + timedelta(
        seconds=random.randint(0, int((end_date - start_date).total_seconds())))

descricao_opcoes = [
    "Lisboa Bistro", "Porto Gourmet", "Coimbra Café",
    "Mercado Central", "Compras Diárias", "Alimentos Frescos",
    "Cinema do Bairro", "Parque Aventura", "Arte e Cultura",
    "Loja de Roupas", "Eletrônicos Modernos", "Livraria Cultural"
]

palavras_chave_para_categorias = {
    "Restaurante": "Restaurante",
    "Café": "Café",
    "Supermercado": "Supermercado",
    "Alimentos": "Supermercado",
    "Mercado": "Supermercado",
    "Lazer": "Lazer",
    "Compras": "Compras",
    "Eletrônicos": "Compras",
    "Loja de Roupas": "Compras",
    "Livraria": "Compras",
    "Arte": "Lazer",
    "Cinema": "Lazer",
    "Parque": "Lazer",
    "Viagem": "Lazer",
    "Museu": "Lazer",
    "Transporte": "Lazer",
    "Gasolina": "Lazer",
    "Hotel": "Lazer",
    "Passagem": "Lazer",
    "Educação": "Educação",
    "Escola": "Educação",
    "Livros": "Educação",
    "Taxa": "Taxa",
    "Juros": "Taxa",
    "Tarifa": "Taxa",
    "Seguro": "Seguro",
    "Saúde": "Saúde",
    "Hospital": "Saúde",
    "Médico": "Saúde",
    "Farmácia": "Saúde",
    "Doação": "Doação",
    "Caridade": "Doação",
    "Presente": "Presente",
    "Transferência": "Transferência",
    "Salário": "Salário",
    "Fatura": "Fatura",
    "Pagamento": "Pagamento",
    "Imposto": "Imposto",
    "Aluguel": "Aluguel",
    "Moradia": "Moradia",
    "Serviços": "Serviços",
    "Internet": "Serviços",
    "Televisão": "Serviços",
    "Água": "Serviços",
    "Eletricidade": "Serviços",
    "Bistro": "Restaurante",
    "Gourmet": "Restaurante"
}

def obter_categoria(descricao):
    # Verificar se alguma palavra-chave está presente na descrição
    for palavra_chave, categoria in palavras_chave_para_categorias.items():
        if palavra_chave in descricao:
            return categoria
    # Se nenhuma palavra-chave corresponder, categorizar como "Outro"
    return "Outro"

def gerar_extratos(mes, ano):
    extratos = []

    # Adicionar uma ou duas transações de Salário com valores aleatórios entre 1000.0 e 3000.0
    for _ in range(random.randint(1, 2)):
        data_salario = random_date(ano, mes, 15).strftime('%d/%m/%Y')
        valor_salario = round(random.uniform(1000.0, 3000.0), 2)
        extratos.append([data_salario, valor_salario, "Salário", "Salário"])

    # Gerar transações de débito
    for _ in range(random.randint(30, 40)):
        data = random_date(ano, mes, 1).strftime('%d/%m/%Y')
        valor = round(random.uniform(-150.0, -5.0), 2)  # Valores negativos representam débitos
        descricao = random.choice(descricao_opcoes)
        categoria = obter_categoria(descricao)
        extratos.append([data, valor, descricao, categoria])

    with open(f'extratos_ficticios_{mes}_{ano}.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Data", "Valor", "Descrição", "Categoria"])
        writer.writerows(extratos)

gerar_extratos(10, 2023)
```
